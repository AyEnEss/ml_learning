```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
import os
from sklearn import datasets

diabetes = datasets.load_diabetes()
```

```python
diabetes.keys()
```

```python
print(diabetes.DESCR)
```

```python
diabetes_df = pd.DataFrame(diabetes.data)
```

```python
np.array(diabetes_df)
```

```python
diabetes_df.columns = diabetes.feature_names
```

```python
diabetes_df['target'] = diabetes.target
```

```python
diabetes_target = diabetes.target
```

```python
diabetes_df.hist(bins=50, figsize=(20,15))
plt.show()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(diabetes_df, diabetes_target, test_size=0.2, random_state=42)
```

```python
print('The training set contains ', len(X_train))
print('The testing set contains ', len(X_test))
```
